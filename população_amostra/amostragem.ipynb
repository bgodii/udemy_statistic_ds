#### Amostragem Aleatória Simples

```python
df_amostra_aleatoria_simples = df.sample(n = 100, random_state=0) # n = quantidade de elementos, random_state = seed
df_amostra_aleatoria_simples.head()
```

```python
def amostragem_aleatoria_simples(dataset: pd.DataFrame, amostra: int, seed: int = False) -> pd.DataFrame:
    if seed:
        return dataset.sample(n = amostra, random_state=seed)
    return dataset.sample(n = amostra)
```

```python
df_amostra_aleatoria_simples = amostragem_aleatoria_simples(df, 100)
df_amostra_aleatoria_simples.head(3)
```
